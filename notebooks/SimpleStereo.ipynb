```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
imgL = cv.imread('../data/im0.png',0)
imgR = cv.imread('../data/im1.png',0)
stereo = cv.StereoBM_create(numDisparities=176, blockSize=25)
disparity = stereo.compute(imgL,imgR)
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(cv.cvtColor(imgL, cv.COLOR_BGR2RGB))
axarr[1].imshow(disparity,'gray')
plt.show()
```
